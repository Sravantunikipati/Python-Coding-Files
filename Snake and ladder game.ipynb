```python
import random

snakes = {16: 6, 47: 26, 49: 11, 56: 53, 62: 19, 64: 60, 87: 24, 93: 73, 95: 75, 98: 78}
ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 36: 44, 51: 67, 71: 91, 80: 100}

player1_name = input('Enter your name : ')
player2_name = input('Enter your name : ')
print("{} vs {}".format(player1_name,player2_name))
player1_pos = 1
player2_pos = 1

current_player = player1_name or player2_name

while player1_pos < 100 and player2_pos < 100:
    dice_roll = random.randint(1, 6)
    input("{}'s turn. Press Enter to roll the dice...".format(player1_name))
    print("You rolled a", dice_roll)
    
    if current_player == player1_name:
        new_pos = player1_pos + dice_roll
        if new_pos in snakes:
            new_pos = snakes[new_pos]
            print('Oops! You got a snake bite')
            print('{} now your position is at {}'.format(player1_name,new_pos))
        elif new_pos in ladders:
            new_pos = ladders[new_pos]
            print('Wow! You climbed a ladder')
            print('{} now your position is at {} , keep going'.format(player1_name,new_pos))
        player1_pos = new_pos
    
    else:
        player2_pos = player2_pos + dice_roll
    
    dice_roll = random.randint(1,6)
    input("{}'s turn. Press Enter to roll the dice...".format(player2_name)) 
    print("You rolled a", dice_roll)
    player2_pos+=dice_roll
    
    if current_player == player2_name:
        new_pos =  player2_pos + dice_roll
        if new_pos in snakes:
            new_pos = snakes[new_pos]
            print('Oops! You got a snake bite')
            print('{} now your position is at {}'.format(player2_name,new_pos))
        elif new_pos in ladders:
            new_pos = ladders[new_pos]
            print('Wow! You climbed a ladder')
            print('{} now your position is at {} , keep going'.format(player2_name,new_pos))
        player2_pos = new_pos

    print("{}'s position : {} and {}'s position : {}".format(player1_name,player1_pos,player2_name,player2_pos))
    
    if player1_pos == 100:
        print("Congrajulations {} you have won!".format(player1_name))
        break
    elif player2_pos == 100:
        print("Congrajulations {} you have won !".format(player2_name))
        break       
print("Game Over")
```
